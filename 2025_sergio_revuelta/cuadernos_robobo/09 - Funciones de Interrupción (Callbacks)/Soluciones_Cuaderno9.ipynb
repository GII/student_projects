```python
from robobopy.Robobo import Robobo

#Funci贸n que se ejecuta cuando el callback se activa
def fun():
    if rob.readTapSensor().zone=="eye":
        rob.moveWheels(20,20)
    elif rob.readTapSensor().zone=="right":
        rob.moveWheels(5,-5)
    elif rob.readTapSensor().zone=="left":
        rob.moveWheels(-5,5)
    elif rob.readTapSensor().zone=="chin":
        rob.moveWheels(40,40)
    elif rob.readTapSensor().zone=="forehead":
        rob.moveWheels(-20,-20)
    elif rob.readTapSensor().zone=="mouth":
        rob.stopMotors()
    
rob = Robobo('localhost')
rob.connect()

rob.moveWheels(30,30)
rob.whenATapIsDetected(fun)
timeout=0
# el programa se mantiene en ejecuci贸n continua durante medio minuto
while timeout<300:
    rob.wait(0.1)
    timeout+=1

from robobopy.Robobo import Robobo
from robobopy.utils.Wheels import Wheels
rob = Robobo('localhost')
rob.connect()
parar = 0
while parar<300:
    rob.wait(0.1)
    parar=parar+1

rob.stopMotors()
rob.disconnect()
```

```python
from robobopy.Robobo import Robobo
from robobopy.utils.BlobColor import BlobColor
from robobopy.utils.Wheels import Wheels
from robobopy.utils.IR import IR

rob = Robobo('localhost')
rob.connect()
rob.moveTiltTo(90,15)
def Motor_Colores():
    if rob.readColorBlob(BlobColor.RED).size>25 and rob.readColorBlob(BlobColor.RED).posx>=50:
        rob.stopMotors()
        rob.wait(1)
        rob.moveWheels(-5,5)
        rob.wait(1)
    elif rob.readColorBlob(BlobColor.GREEN).size>25 and rob.readColorBlob(BlobColor.GREEN).posx>=50:
        rob.moveWheels(-7,7)
    elif rob.readColorBlob(BlobColor.CUSTOM).size>25 and rob.readColorBlob(BlobColor.CUSTOM).posx>=50:
        rob.moveWheels(-3,3)
    elif rob.readColorBlob(BlobColor.BLUE).size>25 and rob.readColorBlob(BlobColor.BLUE).posx>=50:
        rob.moveWheelsByTime(-2,2,2)
        rob.moveWheels(-10,10)


rob.moveWheels(-5,5)
timeout=0
rob.whenANewColorBlobIsDetected(Motor_Colores)


while timeout<40:
    rob.wait(0.5)
    timeout+=1
rob.stopMotors()

rob.disconnect()
```

```python
from robobopy.Robobo import Robobo

#Funci贸n que se ejecuta cuando el callback se activa
def f_callback():
    fling_angle = rob.readFlingAngle()
    print(f" Angulo: {fling_angle}")
    if fling_angle<20 or fling_angle>340:
        rob.moveWheels(15,-15)
    elif fling_angle>150 and fling_angle<210:
        rob.moveWheels(-15,15)
    elif fling_angle>70 and fling_angle<110:
        rob.moveWheels(20,20)
    elif fling_angle>250 and fling_angle<290:
        rob.moveWheelsByTime(-5,-5,2)

rob = Robobo('localhost')
rob.connect()

rob.whenAFlingIsDetected(f_callback)

tiempo=0
# el programa se mantiene en ejecuci贸n continua hasta que pase 1 minuto
while tiempo<450:
    rob.wait(0.1)
    tiempo+=1

print("TIMEOUT")
rob.stopMotors()
rob.disconnect()
```
