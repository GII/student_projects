```python
#Ej 1

from robobopy.Robobo import Robobo
from robobopy.utils.Wheels import Wheels
rob = Robobo('localhost')
rob.connect()
parar = 0
while parar < 60:
    fling_angle = rob.readFlingAngle()
    fling_distance=rob.readFlingDistance()
    print(f" Angulo: {fling_angle} //  Distancia: {fling_distance}")

    if fling_distance>40:
        if fling_angle<45 or fling_angle>315:
            rob.moveWheels(5,-5)
        elif fling_angle>135 and fling_angle<225:
            rob.moveWheels(-5,5)
        else:
            rob.stopMotors()
    rob.wait(1)
    parar=parar+1

rob.stopMotors()
```

```python
#Ej 2
"""
- Si tocas la zona *eye*, el robot se moverá a una velocidad de 20 de forma constante.
- Si tocas la zona *right*, el robot girará hacia la izquierda a una velocidad de 5.
- Si tocas la zona *left*, el robot girará hacia la derecha a una velocidad de 5.
- Si tocas la zona *chin*, el robot avanzará el doble de rápido, es decir, a 40.
- Si tocas la zona *forehead*, el robot avanzará la mitad de rápido , es decir, a 10.
- Si tocas la zona *mouth*, el robot se detendrá.

"""



from robobopy.Robobo import Robobo
from robobopy.utils.Wheels import Wheels
rob = Robobo('localhost')
rob.connect()
parar = 0
while parar < 60:
    tap = rob.readTapSensor()
    print(f" Zona: {tap.zone}")

    if tap.zone=="eye":
        rob.moveWheels(20,20)
    elif tap.zone=="right":
        rob.moveWheels(5,-5)
    elif tap.zone=="left":
        rob.moveWheels(-5,5)
    elif tap.zone=="chin":
        rob.moveWheels(40,40)
    elif tap.zone=="forehead":
        rob.moveWheels(10,10)
    elif tap.zone=="mouth":
        rob.stopMotors()
    rob.wait(1)
    parar=parar+1

rob.stopMotors()
```
