EJERCICIO 1

```python
from robobopy.Robobo import Robobo
from robobopy.utils.Wheels import Wheels
from robobopy.utils.IR import IR


rob = Robobo('localhost')
rob.connect()


rob.startQrTracking()

#Se acerca al vaso de papel para tenerlo dentro de las pinzas
rob.moveWheels(10,10)
while rob.readIRSensor(IR.FrontC)<550:
    rob.wait(0.2)


#Se pone a buscar el QR del papel
rob.stopMotors()
rob.moveTiltTo(75,15)
rob.moveWheels(-2,2)
while rob.readQR().id!="PAPER":
    rob.wait(0.2)


#Se aproxima al depósito del papel
rob.stopMotors()
rob.sayText("Ahí está el depósito del papel")
rob.moveWheels(15,15)

while rob.readQR().id=="PAPER" and rob.readQR().distance<55:
    print(rob.readQR().distance)
    rob.wait(0.2)

#Se separa del vaso de papel y termina el programa
rob.moveWheelsByTime(-15,-15,2)
rob.sayText("Ya he dejado el vaso en su sitio")

rob.stopQrTracking()
rob.disconnect()

```

```python
from robobopy.Robobo import Robobo
from robobopy.utils.Wheels import Wheels
from robobopy.utils.BlobColor import BlobColor
from robobopy.utils.IR import IR

rob = Robobo('localhost')
rob.connect()

#Se le pregunta si quiere aparcar en el puesto 1 o en el puesto 2
Parking=input("¿Dónde aparcará Robobo? ¿En el puesto 1 o en el puesto 2?")
rob.startQrTracking()

#Se mueve el Pan para poder leer el QR bien
rob.movePanTo(-90,35)

#Se mueve lentamente mirando los QRs
rob.moveWheels(6,6)

#Se detendrá cuando el QR sea el indicado y el código esté en la esquina derecha de la pantalla
while not rob.readQR().id == Parking or rob.readQR().x < 275:
    rob.wait(0.25)
rob.stopMotors()

#Se mueve el Pan a su posición original
rob.movePanTo(0,45)

#Primer movimiento
rob.moveWheelsByTime(-25,-10,2.2)
#Segundo movimiento
rob.moveWheelsByTime(12,-21,1)

#Se acerca al poste del parking
rob.moveWheels(4,4)

while rob.readIRSensor(IR.FrontC)<100:
    rob.wait(0.25)


#Se detiene y finaliza el programa
rob.stopMotors()

rob.disconnect()
```

```python
from robobopy.Robobo import Robobo
from robobopy.utils.Wheels import Wheels

rob = Robobo('localhost')
rob.connect()

rob.startQrTracking()
rob.movePanTo(10,10)

#Se declaran las variables
ID="Null"
V=0

#No se detendrá hasta que detecte la señal de stop
while ID!="stop":

    #Condiciones para leer el ID: Mínimo la distancia debe de ser de 15 y el ID no debe ser el mismo que se leyó antes
    #para evitar repetir el código (Esto se podrá arreglar cuando aprendan los callbacks de Robobo)
    if rob.readQR().distance>15 and ID!=rob.readQR().id:
        ID=rob.readQR().id
        print(ID)

        #Velocidad de los motores
        if ID=="10" or ID=="30":
            V=int(ID)
            rob.moveWheels(V,V)
            rob.wait(0.3)
        #Se detiene mientras avanza el peatón
        elif ID=="peatones":
            rob.stopMotors()
            rob.wait(5)
            rob.moveWheels(V,V)
            
rob.stopQrTracking()
rob.stopMotors()
```
